### Membuat Formula untuk penggunaan Central Differences dan Normalisasi L2

```python
def centDifAndNorm(img, h):
    img = img.astype(float)
    #Central difference based on x axis and call it as dx.
    dx = (img[h:-h, 2*h:] - img[h:-h, 0:-2*h]) / (2*h)
    #Central difference based on y axis and call it as dy.
    dy = (img[2*h:, h:-h] - img[0:-2*h, h:-h]) / (2*h)
    
    img2 = []
    img2.append(dx)
    img2.append(dy)
    
    #L2 Norm
    l2 = np.sqrt(dx**2 + dy**2)    
    img2.append(l2)
    
    return img2
```

```python
res_img = centDifAndNorm(image, 1)
```
